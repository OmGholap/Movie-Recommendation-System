```python
def convert(obj):
  l = list()
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
def convert_top3(obj):
  l = list()
  count = 0
  for i in ast.literal_eval(obj):
    if count != 3:
      l.append(i['name'])
      count+=1
  return l
```

```python
def overview(x):
  x = str(x)
  words = nltk.word_tokenize(x)
  words = [word for word in words if word not in string.punctuation]
  filtered_words = [word for word in words if word.lower() not in stopwords.words('english')]

  filtered_words_list = []

  for word in filtered_words:
      filtered_words_list.append(word)
  return filtered_words_list
```

```python
movies_db['overview'] = movies_db['overview'].apply(overview)
```

```python
movies_db['tags'] = movies_db['overview'] + movies_db['genres'] + movies_db['keywords'] + movies_db['tagline'] + movies_db['cast'] + movies_db['crew']
```

```python
prepro_movies_db = movies_db[['movie_id', 'title', 'tags']]
```

```python
prepro_movies_db.head()
```

```python
prepro_movies_db['tags'] = prepro_movies_db['tags'].apply(lambda x : " ".join(x))
```

```python
prepro_movies_db.iloc[0].tags
```

```python
prepro_movies_db['tags'] = prepro_movies_db['tags'].apply(lambda x :x.lower())
```

```python
# vectorization

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(prepro_movies_db['tags']).toarray()
```

```python
vectors[0]
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# stemming helper functino

def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
prepro_movies_db['tags'] = prepro_movies_db['tags'].apply(stem)
```

```python
prepro_movies_db.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
def recommend(movie):
  movie_index = prepro_movies_db[prepro_movies_db['title']== movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1]) [1:6]

  for i in movies_list:
    print(prepro_movies_db.iloc[i[0]].title)
```
